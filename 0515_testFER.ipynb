```python
# セル1：ライブラリの読み込み
from fer import FER
import cv2
import matplotlib.pyplot as plt

# Jupyter内で画像を表示するための設定
%matplotlib inline
```

```python
# セル2：画像を読み込んで表示（画像はこのNotebookと同じフォルダに置く）
image_path = "fearface.jpg"
# image_path = "face1.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis("off")
plt.title("入力画像")
plt.show()
```

```python
# セル3：表情分析してストレススコアを出力
detector = FER(mtcnn=True)
results = detector.detect_emotions(image_rgb)

if results:
    emotions = results[0]["emotions"]
    print("表情スコア:", emotions)

    # 簡易的なストレス指標（怒り＋嫌悪＋悲しみ）
    stress_score = emotions["angry"] + emotions["disgust"] + emotions["sad"]
    print(f"推定ストレススコア: {stress_score:.2f}（0〜1の範囲）")
else:
    print("顔が検出されませんでした。")
```
